```python
import seaborn as sns
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file_path = "Big_Monthly_Sales_Dataset.csv"
df = pd.read_csv(file_path)
```

```python
monthly_sales = df.groupby(["Month_Number", "Month"])["Sales_Amount"].sum().reset_index()
```

```python
monthly_sales = monthly_sales.sort_values("Month_Number")
```

```python
plt.figure(figsize=(10, 5))
plt.bar(monthly_sales["Month"], monthly_sales["Sales_Amount"])
plt.xlabel("Month")
plt.ylabel("Total Sales")
plt.title("Bar Chart - Monthly Total Sales")

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.plot(monthly_sales["Month"], monthly_sales["Sales_Amount"], marker='o')
plt.xlabel("Month")
plt.ylabel("Total Sales")
plt.title("Line Chart - Monthly Total Sales")

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 8))
plt.pie(
    monthly_sales["Sales_Amount"],
    labels=monthly_sales["Month"],
    autopct="%1.1f%%"
)
plt.title("Pie Chart - Monthly Sales ")
plt.show()
```
