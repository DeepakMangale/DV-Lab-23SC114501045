```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Big_Monthly_Sales_Dataset.csv")
```

```python
monthly_sales = df.groupby(["Month_Number", "Month"])["Sales_Amount"].sum().reset_index()
monthly_sales = monthly_sales.sort_values("Month_Number")
```

```python
plt.figure(figsize=(13,5))
plt.bar(monthly_sales["Month"], monthly_sales["Sales_Amount"])
plt.title("Default Color Palette")

plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.bar(monthly_sales["Month"], monthly_sales["Sales_Amount"], color="black")
plt.title("High Contrast (Black)")

plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.barplot(x="Month", y="Sales_Amount", data=monthly_sales, palette="viridis")
plt.title("Viridis Color Palette")

plt.show()
```
